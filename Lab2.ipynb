```python
# 2. Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
# Lọc nam
df_male = df[df["GT"] == "M"]
# df_male

sorted_value_DH2_male = df_male.sort_values(by="DH2")
display(sorted_value_DH2_male)

# Lọc nữ
df_female = df[df["GT"] == "F"]
sorted_value_DH2_female = df_female.sort_values(by="DH2")
display(sorted_value_DH2_female)
```

```python
sorted_value_DH2 = df.sort_values(by=["GT", "DH2"])
sorted_value_DH2["GT"]
```

```python
def Q1(g):
    return np.percentile(g, 25)

def Q2(g):
    return np.percentile(g, 50)

def Q3(g):
    return np.percentile(g, 75)
```

```python
#  Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, 
# Q2 và Q3 của DH1 theo KT

pivot = df.pivot_table(values= "DH1", index="KT", aggfunc=["count", "sum", "mean", "median", "min", "max", "std", Q1, Q2, Q3])
pivot
```

```python
# Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT và KV

pivot2 = df.pivot_table(values= "DH1", index=["KT", "KV"], aggfunc=["count", "sum", "mean", "median", "min", "max", "std", Q1, Q2, Q3])
pivot2
```

```python
# Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT, KV và DT
pivot3 = df.pivot_table(values= "DH1", index=["KT", "KV", "DT"], aggfunc=["count", "sum", "mean", "median", "min", "max", "std", Q1, Q2, Q3])
pivot3
```

```python
# Hãy trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thỏa mãn điều kiện (DH1 >= 5.0 và DH2 >= 4.0 và DH3 >= 4.0)

df_M_grade = df[(df['KV'] == 1) & (df['GT'] == 'M') & (df["DH1"] >= 5) & (df["DH2"] >= 4) & (df["DH3"] >= 4)].count()
df_M_grade
```
