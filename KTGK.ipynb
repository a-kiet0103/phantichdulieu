```python
#CÂU 3:
#Câu a:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
import statsmodels.formula.api as smf
from scipy import stats
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Câu b:
df=pd.DataFrame({'Company':['AP','ASM','DR','FM','GL','IHF','PRT','Ray','Real','SS'],
'Selling': [5,9,6.7,8.75,3,13.6,4.6,6.7,3,7.7],
'Price':[15,14,15,17,11,19,13,14,10,13] })

print("The original dataframe:\n")
print(df)
```

```python
plt.plot(df['Selling'],df['Price'], 'o',linewidth=2);
plt.xlabel('Postmost size')
plt.ylabel('Premolt size')
plt.title('Postmost vs Premolt')
```

```python
pearsonr(df['Selling'],df['Price'])
```

```python
result = smf.ols('Selling~Price', df).fit()
print('DU  BAO', result.summary())
```

```python
#y dự báo dựa vào phương trình hồi quy
sales_pred = result.predict() # dự đoán
plt.plot(df['Selling'],df['Price'],'o', color = 'b') 
plt.plot(df['Selling'],sales_pred,'r',linewidth=2) 
plt.xlabel('Postmolt size')
plt.ylabel('Premolt size')
plt.title('Premolt and Postmolt ')
plt.show()
```
